* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
purchase_data.columns
```

## Player Count

* Display the total number of players


```python
player_total = len(purchase_data["SN"].unique())
player_total
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# FIND UNIQUE PURCHASES
unique_items = len(purchase_data["Item ID"].unique())
unique_items

#CALCUATE THE AVERAGE
average_price = purchase_data["Price"].mean()
average_price

# CALCULATE THE TOTAL REVENUE 
total_revenue = purchase_data["Price"].sum()
```

```python
summary_table = pd.DataFrame({"Unique Items Count": [unique_items],
                              "Average Price": [average_price],
                              "Total Purchases": purchase_data["SN"].count(),
                              "Total Revenue": [total_revenue]})

summary_table["Average Price"] = summary_table["Average Price"].map("${:.2f}".format)
summary_table["Total Revenue"] = summary_table["Total Revenue"].map("${:.2f}".format)
summary_table
```

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
drop_dupes_df = purchase_data.drop_duplicates("SN")
gender_count = drop_dupes_df.groupby("Gender").count()["Purchase ID"]
gender_percent = gender_count/player_total*100
gender_df = pd.DataFrame({'Percentage of Players': gender_percent})
gender_df["Purchase Count"] = gender_count

gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
genders = purchase_data.groupby("Gender")
count = genders["Gender"].count()

#CALCUATE THE AVERAGE
gender_price = genders["Price"].mean()

# CALCULATE THE TOTAL REVENUE 
gender_revenue = genders["Price"].sum()
```

```python
genders = pd.DataFrame({"Average Price": gender_price,
                        "Total Purchases": count,
                        "Total Revenue": gender_revenue})

genders["Average Price"] = genders["Average Price"].map("${:.2f}".format)
genders["Total Revenue"] = genders["Total Revenue"].map("${:.2f}".format)
genders
purchase_data.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
# Sort de bins
age_group = pd.cut(drop_dupes_df["Age"], age_bins, labels=group_names)

# Count the age groups 
age_groups_count = age_group.value_counts(sort=False)

# Calculate the percentage
age_percent = round((age_groups_count/player_total) * 100, 2)

# Make a dataframe 
age_df = pd.DataFrame({
                    "Percentage of Players": age_percent,
                    "Age Count": age_groups_count})

age_df["Percentage of Players"] = age_df["Percentage of Players"].map("{:.2f}%".format)
age_df
purchase_data.head()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting




* Display the summary data frame

```python
#Making a copy of the OG data set cus it makes me feel better
age_group_df = purchase_data


# Creating the bins 
age_group2 = pd.cut(age_group_df["Age"], age_bins, labels=group_names)

# Creating the index "AGE GROUP"
age_group_df ["Age Groups"] = age_group2
age_group_df




# Dropin the dupes cus I think we should
drop_dupes_df2 = age_group_df.drop_duplicates("SN")

age_groupy = age_group_df.groupby("Age Groups")
age_groupy

# count_makers = maker_group['maker'].count()

#PURCHASING ANALYSIS 
purchase_count = round(age_groupy["Price"].count(),2)

# AVERAGE PURCHASE PRICE 
avg_price = round(age_groupy["Price"].mean(),2)

# TOTAL PURCHASE VALUES
total_purchases = round(age_groupy["Price"].sum(), 2)

# "NORMALIZED" â€“ average cost that each user spends
total_users_num = drop_dupes_df2.groupby("Age Groups").count()["SN"]

age_groupy_df = pd.DataFrame({
                    "Purchase Count": purchase_count,
                    "Average Prices": avg_price,
                    "Total Purchase": total_purchases,
                    "Normalized Totals": round(total_purchases/total_users_num, 2)})
age_groupy_df

#Format to $
for i in age_groupy_df.columns[1:]:
    age_groupy_df[i] = age_groupy_df[i].astype(float).map("${:}".format)
    
age_groupy_df
# So when running this age_group shows up in the DataFrame, why??? 
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# TOP SPENDERS
top_spender_groupby = purchase_data.groupby(["SN"])

# TOP SPENDERS COUNT
top_spender_count = top_spender_groupby.count()["Purchase ID"]

# AVERAGE PRICE 
top_spender_avg = round(top_spender_groupby.mean()["Price"],2)
#top_spender_avg

# TOTAL SPENDING
top_spender_tots = round(top_spender_groupby.sum()["Price"],2)

top_spender_df = pd.DataFrame({
                    "Purchase Count": top_spender_count,
                    "Average Purchase Price": top_spender_avg,
                    "Total Purchase": top_spender_tots})

top_spender_df = top_spender_df.sort_values(by= ["Total Purchase"], ascending = False)

top_spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grouby_items = purchase_data.groupby(["Item ID", "Item Name"])

item_count = round(grouby_items["Price"].count(),2)

item_total = round(grouby_items["Price"].sum(),2)

item_price = round(item_total / item_count,2)

item_df = pd.DataFrame({
            "Purchase Count": item_count,
            "Item Name": item_price ,
            "Total Purchase Value": item_total
})

item_df = item_df.sort_values(by= ["Purchase Count"], ascending = False)

#Make it money
for i in item_df.columns[1:]:
    item_df[i] = item_df[i].astype(float).map("${:}".format)

item_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profit_df = pd.DataFrame({
            "Purchase Count": item_count,
            "Item Name": item_price ,
            "Total Purchase Value": item_total
})

profit_df = profit_df.sort_values(by= ["Total Purchase Value"], ascending = False)

profit_df.head()
```
